```
import random
import pandas as pd

def random_walk_3D(n,speed=1):
  result = {
      'steps' : [0],
      'x' : [0],
      'y' : [0],
      'z' : [0]
  }
  for step in range(n):
    result['steps'].append(step)
    result['x'].append(result['x'][-1] + random.uniform(-speed,speed))
    result['y'].append(result['y'][-1] + random.uniform(-speed,speed))
    result['z'].append(result['z'][-1] + random.uniform(-speed,speed))
  return result
```

```
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import linear_cmap
from bokeh.models import ColumnDataSource
from bokeh.palettes import Viridis256
import pandas as pd

output_notebook()

df = pd.DataFrame.from_dict(random_walk_3D(1000, speed=5))
source = ColumnDataSource(df)

mapper = linear_cmap(field_name='steps',low=min(df.steps) ,high=max(df.steps), palette=Viridis256)

p = figure()
p.line(x='x', y='y', source=source, color='lightgrey')
p.circle(x='x', y='y', color=mapper, source=source,size=5, alpha=0.5)

show(p)
```
